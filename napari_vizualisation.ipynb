# Visualiaztion with Napari

```python
import numpy as np
import nibabel as nib
# import nilearn as nl
import nilearn.plotting as nplt
import os
import napari
import matplotlib.pyplot as plt
```

## Some file stats

```python
img_path = "data/Patient test 2-liver_post_ablation_CORRIGE.nii.gz"

img = nib.load(img_path)

print(f"Dimensions : {img.shape}")

# 2. Résolution des voxels en mm
print(f"Résolution (mm) : {img.header.get_zooms()}")

# 3. Type de données
print(f"Type de données : {img.get_data_dtype()}")

orientation = nib.aff2axcodes(img.affine)
print(f"Orientation : {orientation}") 
```

## Patient test 1 liver & ablation

```python
# 1. Charger l'image du patient
img_patient = nib.load('data/Patient test 1-liver post ablation_CORRIGE.nii.gz')
data_patient = img_patient.get_fdata()

# 2. Charger le masque de la zone d'ablation
img_mask = nib.load('data/Patient test 1-ablation zone_CORRIGE.nii.gz')
data_mask = img_mask.get_fdata()

# 3. Lancer le viewer
viewer = napari.Viewer()

# Ajouter l'image (en ajustant le contraste automatiquement)
viewer.add_image(data_patient, name='Foie 1', colormap='gray')

# Ajouter le masque (en le mettant en rouge transparent)
viewer.add_labels(data_mask.astype(int), name='Zone Ablation', opacity=0.5)

napari.run()
```

## Patient test 2 liver & ablation

```python
# 1. Charger l'image du patient
img_patient = nib.load('data/Patient test 2-liver_post_ablation_CORRIGE.nii.gz')
data_patient = img_patient.get_fdata()

# 2. Charger le masque de la zone d'ablation
img_mask_1 = nib.load('data/Patient test 2-ablation_zone_1_CORRIGE.nii.gz')
data_mask_1 = img_mask_1.get_fdata()

img_mask_2 = nib.load('data/Patient test 2-ablation_zone_2_CORRIGE.nii.gz')
data_mask_2 = img_mask_2.get_fdata()

# 3. Lancer le viewer
viewer = napari.Viewer()

# Ajouter l'image (en ajustant le contraste automatiquement)
viewer.add_image(data_patient, name='Foie 2', colormap='gray')

# Ajouter le masque (en le mettant en rouge transparent)
viewer.add_labels(data_mask_1.astype(int), name='Zone Ablation 1', opacity=0.5)
viewer.add_labels(data_mask_2.astype(int), name='Zone Ablation 2', opacity=0.5)

napari.run()
```
